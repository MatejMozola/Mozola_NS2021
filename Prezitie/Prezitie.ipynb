```python
#Stiahnutie datasetu
!pip install wget
import wget
url = 'https://drive.google.com/file/d/14xBPfFz-sX1FxN1TCp0wFtXussNn-WID/view?usp=sharing'
wget.download(url, '/test_data.csv')
url = 'https://drive.google.com/file/d/1YRj_uz1hyATF6hRB5zudJKQ4vmsWkTNF/view?usp=sharing'
wget.download(url, '/train_data.csv')

#Načítanie datasetu
import pandas as pd
new_header = ['id','id_pasaziera','prezil','pohlavie','vek','cestovne',
              'prva_trieda','druha_trieda','tretia_trieda','velkost_rodiny',
              'zenaty','vydata','slobodny','slobodna','vycestovanie1',
              'vycestovanie2','vycestovanie3']
train = pd.read_csv("/train_data.csv", header=None, skiprows=1, names=new_header)
test = pd.read_csv('/test_data.csv', header=None, skiprows=1, names=new_header)

#Odstránenie nadbytočných dát
train.drop(['id','id_pasaziera'], axis=1)

test.drop(['id','id_pasaziera'], axis=1)

#Rozdelenie dát na tréning a test
atributy = ['pohlavie','vek','cestovne','prva_trieda','druha_trieda',
            'tretia_trieda','velkost_rodiny','zenaty','vydata','slobodny',
            'slobodna','vycestovanie1','vycestovanie2','vycestovanie3']

X_train = train[atributy]
y_train = train.prezil
X_test = test[atributy]
y_test = test.prezil

#Výpis formátu dát
print("Formát dát")
print("Tréningové:",X_train.shape,y_train.shape, " Testovacie:",X_test.shape,y_test.shape)

# Import MLP
from sklearn.neural_network import MLPRegressor

# Vytvorenie modelu
regr = MLPRegressor(random_state=1, max_iter=100)

# Trenovanie modelu
regr.fit(X_train, y_train)

# Predpovedanie hodnôt pre testovacie dáta
y_pred = regr.predict(X_test)

# Vypis presnosti modelu
print("Presnosť modelu: ", regr.score(X_test, y_test))

# Ukazka porovnania predpovedanych a realnych hodnot
y_pred_series = pd.Series(y_pred)
pd.concat([y_test, y_pred_series], axis=1)
```
